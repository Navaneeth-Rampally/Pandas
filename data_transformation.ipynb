**2. Value counts** - is one of the quickest ways to explore categorical data. It returns a Series containing counts of unique values, effectively creating a frequency table.

```python
df['Gender'].value_counts()
```

```python
df['Education'].value_counts()
```

**3. Grouping Data**

**GroupBy**- It is a flexible framework (Split-Apply-Combine) that lets you group data and calculate any statistic (sum, mean, max, standard deviation) on other columns.

```python
df.groupby(['Gender'])['ApplicantIncome'].mean()
```

**4. Feature Engineering**- Extracting new columns from existing columns.

```python
df['Valuable'] = "No"
df.loc[(df['ApplicantIncome'] > 2000)& (df['CoapplicantIncome']>5000), 'Valuable'] = "Yes"
```

**PIVOT TABLES**

**Merging Datasets**

```python
students = pd.DataFrame({
    'student_id': [101,102,103,104],
    'name' : ['Ramesh', 'Suresh', 'Mahesh', 'Naresh'],
    'city': ['Hyderabad', 'Vijayawada', 'Guntur', 'Nellore']
})
marks = pd.DataFrame({
    'student_id': [101, 102, 103, 105],
    'subject': ['Maths', 'Maths', 'Science', 'Maths'],
    'marks': [85,78,88,90]
})
```

```python
students.head()
```

```python
marks.head()
```

```python
merged_df = pd.merge(students, marks, on = 'student_id', how = 'inner')
```

```python
merged_df
```

```python
pd.merge(students, marks, on = 'student_id', how = 'outer')
```

```python
merged_file = merged_df.to_csv(r"D:\CPSIT\Practice\Pandas\merged_data.csv")
```
